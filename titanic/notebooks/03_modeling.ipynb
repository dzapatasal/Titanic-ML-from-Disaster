# Carga y preparacion de X e y

```python
# Aseg√∫rate de usar la ruta correcta a tu archivo curado
path_curated = '../data/processed/train_curated.csv' 
df_model = pd.read_csv(path_curated)

# üí° Importante: Elimina las columnas 'Name' y 'Ticket' si no las eliminaste en el curado
df_model = df_model.drop(columns=['Name', 'Ticket']) 

# Definir Caracter√≠sticas (X) y Variable Objetivo (y)
X = df_model.drop('Survived', axis=1) # El resto de las columnas
y = df_model['Survived'] # La variable a predecir ‚§é
```
